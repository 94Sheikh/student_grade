```python
class StudentGradeTracker:
    def __init__(self):
        self.student = []
        
        
    def add_student(self, name, grade):
        self.student.append((name, grade))
#         print(self.student)
        
    def highest_grade(self):
        return max([grade for name, grade in self.student])
        print (f"Hihest grade: {grade}")

    def lowest_grade(self):
        return min([grade for name, grade in self.student])
        print (f"Lovest grade: {grade}")
    
    
    def average_grade(self):
        grades = [grade for name, grade in self.student]
        Ave_grade = sum(grades)/len(grades)
        return Ave_grade
        print(f"Average grade: {Ave_grade}")
    
grade_tracker = StudentGradeTracker()

Studen_number = int(input("Enter Student Number"))

for i in range(Studen_number):
    name = input(f"Enter Studen Name {i + 1}")
    grade = int(input(f"Enter Your grade {name} "))
    
    grade_tracker.add_student(name,grade)
    
# print(f"Student Name {name}")    
print(f"Highest grade: {grade_tracker.highest_grade()}")
print(f"Lowest grade: {grade_tracker.lowest_grade()}")
print(f"Average grade: {grade_tracker.average_grade()}")
           
        
# student1 = StudentGradeTracker()
# student1.add_student("Zulfiqar", 98)
# student1.highest_grade()
        
    
    
```

```python

import csv

def read_data(file_name):
    data = []
    #return data
    
    with open(file_name, 'r') as f:
        reader = csv.reader(f)
        next(reader)
        for row in reader:
            data.append((int(row[0]), row[1], float(row[2])))
            
    return data


def number_of_coustomers(data):
    coustomer_num = []
    for row in data:
        if row[0] not in coustomer_num:
            coustomer_num.append(row[0])
            
    return len(coustomer_num)

def get_total_revenue(data):
    total = 0
    for row in data:
        total += row[2]
        
    return total

def get_revenue_per_customer(data):
    num_coustomers = number_of_coustomers(data)
    total_revenue = get_total_revenue(data)
    
    return num_coustomers/total_revenue


def get_number_purchase(data):
    
    coustomer_count = {}
    
    for row in data:
        if row[0] not in coustomer_count:
            coustomer_count[row[0]] = 1
        else:
            coustomer_count[row[0]] += 1
            
    count = 0
    
    for coustomer, num_purchase in coustomer_count.items():
        if num_purchase > 1:
            count += 1
            
    return count


def main():
    file_name = "data.csv"
    data = read_data(file_name)
    
    
    num_coustomers = number_of_coustomers(data)
    total_revenue = get_total_revenue(data)
    revenue_per_customer = get_revenue_per_customer(data)
    Number_of_purchase = get_number_purchase(data)
    
    
    print(f"Total Number of coustomer: {num_coustomers}")
    print(f"Total revenue: ${total_revenue}")
    print(f"average revenue per customer: ${revenue_per_customer:.2f}")
    print(f"Number of customers with more than one purchase: {Number_of_purchase}")
    
if __name__ == "__main__":
    main()

    
    
    
    
    
    
    
    

    
# file_name = "data.csv"
# data = read_data(file_name)

# print(data)
        
    

```

```python
import csv

def get_student_info():
    name = input("Enter your Name")
    grade = float(input("Enter your grade"))
    
    return (name, grade)

def get_class_grade(num_student):
    student_info = []
    
    for i in range(num_student):
        print(f"Student {i+1}")
        student_info.append(get_student_info())
        
    return student_info
        
def get_highest_grade(class_grade):
    return max([grade for name, grade in class_grade])

def get_lowest_grade(class_grade):
    return min([grade for name, grade in class_grade])

def get_average_grade(class_grade):
    grades = [grade for name, grade in class_grade]
    
    average_grade = sum(grades) / len(grades)
    
    return average_grade

num_student = int(input("Enter Student Number"))


class_grade = get_class_grade(num_student)

for name, grade in class_grade:
    print(f"Student Name & grade: {name} {grade}") 
    
highest_grade = get_highest_grade(class_grade)

lowest_grade = get_lowest_grade(class_grade)

average_grade = get_average_grade(class_grade)
average_grade = round(average_grade, 2)


print()
#print(f"Student: {name}")
#print(f"Student grade: {class_grade}")
print()
print(f"Student High Grade: {highest_grade}")
print(f"Student Low Grade: {lowest_grade}")
print()
print(f"Average grade: {average_grade}")

# adding data in csv file

with open("classess_grades.csv", "w", newline= "" ) as file:
    file_handler = csv.writer(file)
    
    for name, grade in class_grade:
        file_handler.writerow([name, grade])

#         file_handler.writerow(["Studentn Name", name])
#         file_handler.writerow(["Studentn grade", grade])
    file_handler.writerow(["High Grade", highest_grade])
    file_handler.writerow(["Low Grade", lowest_grade])
    file_handler.writerow(["Average Grade", average_grade])
    
    


    

```
